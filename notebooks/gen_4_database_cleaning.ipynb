# Generation IV Pokémon Database Cleaning

This notebook prepares a Generation IV-focused slice of an expandable relational Pokémon database.

The schema is designed to support all generations eventually, but this first pass focuses on Generation IV Pokémon/game data. Raw data is taken from PokéAPI CSV files, cleaned with pandas, and exported as relational CSV tables ready for SQL/Access import.

The ERD below is generated from the canonical Mermaid source file:

`../erd/mpdb_erd.mmd`

```python
from pathlib import Path
import shutil
import subprocess
from IPython.display import SVG, Markdown, display

# Paths are relative to the notebook location: notebooks/gen_4_database_cleaning.ipynb
ERD_SOURCE = Path("../erd/pokemon_database_erd.mmd")
ERD_OUTPUT = Path("../erd/pokemon_database_erd.svg")

if not ERD_SOURCE.exists():
    display(Markdown(f"❌ ERD source file not found: `{ERD_SOURCE}`"))

elif shutil.which("mmdc") is not None:
    subprocess.run(
        ["mmdc", "-i", str(ERD_SOURCE), "-o", str(ERD_OUTPUT)],
        check=True
    )
    display(SVG(filename=str(ERD_OUTPUT)))

elif ERD_OUTPUT.exists():
    display(Markdown(
        "⚠️ Mermaid CLI is not installed, so the ERD was not regenerated. "
        "Displaying the existing SVG instead."
    ))
    display(SVG(filename=str(ERD_OUTPUT)))

else:
    display(Markdown(
        "❌ Mermaid CLI is not installed and no existing SVG was found.\n\n"
        "Install Mermaid CLI with:\n\n"
        "`npm install -g @mermaid-js/mermaid-cli`"
    ))
```
